```python
#모델 이름 입력으로 넣을 수 있게 하자 !
def model_best_test(n_select,fold,data,name=''):
  model_name =  f"{n_select} soft + {fold} +{name}"

  clf=setup(data=data, target = 'target',session_id=1,log_experiment = True)
  
  # 교차검증으로 가장 좋은 모델 찾기
  best = compare_models(n_select= n_select, fold = fold,
                        cross_validation=True,
                        sort='Accuracy')
  
  # n_select개 모델 soft voting 하기
  blend_4_soft = blend_models(estimator_list=best,method='soft',choose_better=True,verbose=True,fold=fold)

  tuned_top3 = [tune_model(i) for i in best]
  blender = blend_models(tuned_top3)
  stacker = stack_models(tuned_top3)
  best_auc_model = automl(optimize = 'AUC')
  
  final_model = finalize_model(best_auc_model)
  prediction = predict_model(final_model, data=test)

  #교차 검증은 어떻게..?
  
  #cv_score =cross_val_score(final_model,data.iloc[:,:-1],data['target'], cv=fold)
  #model_rank.loc[len(model_rank)] = [model_name,cv_score]
  #prediction = final_model.predict()

  return model_name, prediction
```

```python
class MyError(Exception):
  pass
  
def trans_pre2z(model_name,prediction,z):
    cnt=0
    for k in range(0,2000):
      if z['target'].iloc[k] == prediction['Label'].iloc[k]:
        cnt+=1
    
    #맞으면 저장
    if cnt == 2000:
      filename = f'{model_name}'
      z.to_csv('./' + filename + '.csv')  
    
    #아니면 에러처리
    if cnt!=2000:
      raise MyError()
```

```python
def Pred2File(model_name,prediction):
  z=pd.read_csv('./sample_submission.csv')
  z=z.set_index('id')
  z['target']= prediction['Label']

  #옮기기
  for i in range(1999,1,-1):
    z['target'].iloc[i] = z['target'].iloc[i-1]
  
  z['target'].iloc[0] =1
  z['target'].iloc[1] =0
  try:
    trans_pre2z(model_name,prediction,z)
  except MyError:
    print('제대로 옮겨지지 않았습니다.')  
  
```
